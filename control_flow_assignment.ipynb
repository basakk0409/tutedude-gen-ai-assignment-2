#Task 1 — Discount Rules (if / elif / else)

```python
user_input = input("Enter order amount: ")

# Validate numeric input
if user_input.isdigit():
    order_amount = int(user_input)

    # Determine discount
    if order_amount >= 2000:
        discount = 15
    elif order_amount >= 1500:
        discount = 10
    elif order_amount >= 1000:
        discount = 7
    else:
        discount = 0

    discount_value = order_amount * discount / 100
    subtotal = order_amount - discount_value

    # Optional tax (5%)
    tax = subtotal * 5 / 100
    final_total = subtotal + tax

    print("Original Amount:", order_amount)
    print("Discount Applied:", discount, "%")
    print("Subtotal after discount:", subtotal)
    print("Tax (5%):", tax)
    print("Final Amount:", final_total)

else:
    print("Error: Please enter a valid numeric amount.")
```

#Task 2 — Process Multiple Orders (for loop)

```python
orders = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discounted_orders = 0

print("Order  | Discount% | Final Amount")
print("-----------------------------------")

for amount in orders:

    if amount >= 2000:
        discount = 15
    elif amount >= 1500:
        discount = 10
    elif amount >= 1000:
        discount = 7
    else:
        discount = 0

    discount_value = amount * discount / 100
    final_amount = amount - discount_value

    if discount > 0:
        discounted_orders += 1

    total_revenue += final_amount

    print(amount, " | ", discount, "% | ", final_amount)

print("-----------------------------------")
print("Total Revenue after discount:", total_revenue)
print("Orders with discount:", discounted_orders)
```

#Task 3 — User Menu (while loop + break/continue)

```python
orders = []

while True:
    print("\nMenu")
    print("1 - Add order amount")
    print("2 - Show all orders with totals")
    print("q - Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        value = input("Enter order amount: ")

        if value.isdigit():
            orders.append(int(value))
            print("Order added.")
        else:
            print("Invalid amount.")

    elif choice == "2":

        total = 0

        if len(orders) == 0:
            print("No orders available.")
            continue

        print("Amount | Discount% | Final")

        for amount in orders:

            if amount >= 2000:
                discount = 15
            elif amount >= 1500:
                discount = 10
            elif amount >= 1000:
                discount = 7
            else:
                discount = 0

            final = amount - (amount * discount / 100)
            total += final

            print(amount, "|", discount, "% |", final)

        print("Total Revenue:", total)

    elif choice == "q":
        print("Exiting program...")
        break

    else:
        print("Invalid choice. Try again.")
        continue
```

#Task 4 — Loop Control with break & continue

```python
daily = [200, 150, 0, 400, 50, -1, 300]

total_sales = 0

for day in daily:

    if day == -1:
        print("Corrupted data found. Stopping process.")
        break

    if day == 0:
        print("No sales for the day. Skipping.")
        continue

    total_sales += day
    print("Added:", day, "| Running Total:", total_sales)

print("Final Total Sales:", total_sales)
```
